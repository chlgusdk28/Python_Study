```python
from bs4 import BeautifulSoup
import urllib.request as req

nameList = []
starList = []
titleList = []
movieList = []
urlList = []
avgStarLisg = []

# 다시 생각해보기로 했다. 지금 일단 규격이 다름 
#https://movie.naver.com/movie/bi/mi/point.nhn?code=169240 참고!! 
# 이름이 |  가 붙어서 출력되는 부분 체크해야함 이부분을 제거한 후 비교했음 
#찾는 문자열을 따로 넣고 출력함 
# 만약 박평식이 있는 페이지면 해당하는 평론가 평점을 모두 출력해보도록 한다!!!!



# 하루에천개씩 할 생각!!
for page in range(5000):
    no = str(page+24009)
    url='https://movie.naver.com/movie/bi/mi/point.nhn?code='+no
    res =req.urlopen(url)
    #print(url)
    
    soup = BeautifulSoup(res,'html.parser')
    tests = soup.select("#content > div.article > div.section_group.section_group_frst > div:nth-child(6) > div > div.score140 > div > ul > li > div.score_reple > dl > dd")
  
    search = "박평식"
    avgStarTotal = ""
    
    for test in tests :
        name = test.string.strip().lstrip('| ')
        # 데이터 정리 
        if search in name :
            
            print ("---------------------------------------------")
            stars = soup.select("#content > div.article > div.section_group.section_group_frst > div:nth-child(6) > div > div.score140 > div > ul > li > div.star_score > em")
            names = soup.select("#content > div.article > div.section_group.section_group_frst > div:nth-child(6) > div > div.score140 > div > ul > li > div.score_reple > dl > dd")
            titles = soup.select("#content > div.article > div.section_group.section_group_frst > div:nth-child(6) > div > div.score140 > div > ul > li > div.score_reple > p") 
           
            

                
        
            for star in stars :
                print(star.string)
                starList.append(star.string)
                
            for name in names :
                print(name.string.strip().lstrip('| '))
                nameList.append(name.string.strip().lstrip('| '))     
                
            for title in titles : 
                print(title.string)
                titleList.append(title.string)
                movies = soup.select("#content > div.article > div.mv_info_area > div.mv_info > h3 > a")
                avgStars = soup.select("#netizen_point_tab_inner > em") 
                avgStarTotal = ""
                for movie in movies : 
                    print(movie.string)
                    movieList.append(movie.string)
                    urlList.append(url) 
                    
                for avgStar in avgStars :
                    avgStarTotal+=avgStar.string
                print(avgStarTotal)
                avgStarLisg.append(avgStarTotal)
            
                

                

print(nameList)
print(starList)
print(titleList)
print(movieList)
print(urlList)


# import pandas as pd
# # csv 만들기
# df = pd.DataFrame(data=movieList, columns=['Movie'])
# df['name'] = pd.Series(nameList)
# df['star'] = pd.Series(starList)
# df['content'] = pd.Series(titleList)
# df['URL'] = pd.Series(urlList)
# df['avgStar'] = pd.Series(avgStarLisg)

# df.to_csv("../data/navermovie_new.csv", encoding ="utf-8", index=False)
# pd.read_csv("../data/navermovie_new.csv")
```
